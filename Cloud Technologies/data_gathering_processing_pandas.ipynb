```python
# 名称列表
us = ['tsla_data', 'amzn_data', 'appl_data', 'nflx_data', 'MSFT_data']
cn = ['tx_data', 'mt_data', 'byd_data', 'baba_data']
dollar = ['tsla_data', 'amzn_data', 'appl_data', 'nflx_data', 'MSFT_data', 'tx_data', 'baba_data']
rmb = ['mt_data', 'byd_data']
btc = ['btc_data']

def preprocessing(df, df_name):
    df.index = pd.to_datetime(df.index, utc=True).normalize()
    df.drop(['Dividends', 'Stock Splits'], axis=1, inplace=True)
    df['AVG'] = (df['Open'] + df['High'] + df['Low'] + df['Close']) / 4
    df.drop(['Open', 'High', 'Low', 'Close'], axis=1, inplace=True)

    # 用名字判断分组
    if df_name in us and df_name in dollar:
        df['Group'] = 'US'
    elif df_name in cn and df_name in dollar:
        df['Group'] = 'CN'
    elif df_name in cn and df_name in rmb:
        df['Group'] = 'CN'
        df['AVG'] = df['AVG'] / 7
    else:
        df['Group'] = 'BTC'

# 调用时传入变量名
for df_name in us + cn + btc:
    preprocessing(globals()[df_name], df_name)
```

```python
def filter_by_date_range(df, start_date = "2014-11-26", end_date = "2024-11-26"):
    
     # 确保索引是 DatetimeIndex
    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("DataFrame have to be a type of DatetimeIndex")
    
    # 按索引范围过滤数据
    filtered_df = df.loc[start_date:end_date]
    return filtered_df
```

```python
def calculate_value(df, initial_investment=100):

    # 确保 'AVG' 列存在
    if 'AVG' not in df.columns:
        raise ValueError("DataFrame have to contain column 'AVG' ")
    
    # 初始化 Value 列
    df['Value'] = 0.0

    # 设置第一行的初始投资
    df.loc[df.index[0], 'Value'] = initial_investment

    # 逐行计算 Value 列
    for i in range(1, len(df)):
        previous_value = df.iloc[i - 1]['Value']
        previous_avg = df.iloc[i - 1]['AVG']
        current_avg = df.iloc[i]['AVG']
        df.iloc[i, df.columns.get_loc('Value')] = (previous_value / previous_avg) * current_avg

    return df
```

```python
def merge_datasets_with_index(datasets):

    # 合并多个 DataFrame
    combined_df = pd.concat(datasets, axis=0)
    
    # 确保索引是 DatetimeIndex
    if not isinstance(combined_df.index, pd.DatetimeIndex):
        combined_df.index = pd.to_datetime(combined_df.index)

    # 按索引排序
    combined_df = combined_df.sort_index()

    return combined_df
```

```python
datasets = [btc_data, tsla_data, amzn_data, appl_data, nflx_data, tx_data, mt_data, byd_data, baba_data, MSFT_data]
# 调用函数合并数据集
merged_data = merge_datasets_with_index(datasets)
merged_data.head()
```
